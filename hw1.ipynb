```julia
Pkg.add("Convex")
Pkg.add("ECOS")

using Convex, ECOS, PyPlot
n=100
A = randn(100,100)
b = randn(100,1)
```

```julia
# p = 1
x = Variable(n)
p = minimize(0.5*sum_squares(A*x - b) + 10* norm(x,1))
solve!(p,ECOSSolver())

figure(figsize=(20,5))
r = -0.99:(0.1/50):0.99
plt[:hist](x.value, 50, facecolor = "w")
plot(r,10*abs(r), "k")
axis([-1,1,0,10])
```

```julia
# p = 2

x = Variable(n)
p = minimize(0.5*sum_squares(A*x - b) + 10* norm(x,2))
solve!(p,ECOSSolver())

figure(figsize=(20,5))
r = -0.99:(0.1/50):0.99
plt[:hist](x.value, 50, facecolor = "w")
plot(r,10*r.^2, "k")
axis([-1,1,0,10])
```

```julia
# Dead zone
x = Variable(n)
p = minimize(sum_squares(A*x - b) + sum(max(x-1,0) + max(-x-1,0)))
solve!(p,ECOSSolver())

figure(figsize=(20,5))
r = -2:(0.1/50):2
plt[:hist](x.value, 50, facecolor = "w")
plot(r, 10* (max(r-1,0) + max(-r-1,0)), "k")
axis([-2,2,0,10])
```

```julia
# Log barrier
x = Variable(n)
a = 1
p = minimize(sum_squares(A*x - b) - sum(a^2*log(1-x^2/a)))
solve!(p,ECOSSolver())

figure(figsize=(20,5))
r = -1:(0.1/50):1
plt[:hist](x.value, 50, facecolor = "w")
plot(r, -a^2*log(1-r.^2/a), "k")
axis([-1,1,0,10])
```
